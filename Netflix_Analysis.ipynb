```python
import matplotlib.pyplot as plt

# Plot distribution of content added over time
data['Date'].dt.year.value_counts().sort_index().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of Shows Added')
plt.title('Number of Shows Added Per Year')
plt.show()
```

```python
# Scatter plot of 'Minutes' vs 'release_year'
plt.scatter(data['release_year'], data['Minutes'])
plt.xlabel('Release Year')
plt.ylabel('Minutes')
plt.title('Release Year vs Duration')
plt.show()

# Histogram of 'Minutes'
plt.hist(data['Minutes'], bins=30, edgecolor='k')
plt.xlabel('Minutes')
plt.ylabel('Frequency')
plt.title('Distribution of Duration')
plt.show()
```

```python
# Average duration by type
print(data.groupby('type')['Minutes'].mean())

# Number of shows by country
print(data['country'].value_counts())

# Ratings distribution
plt.figure(figsize=(12,6))
data['rating'].value_counts().plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')
plt.show()
```

```python
import seaborn as sns

# Calculate and plot correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
